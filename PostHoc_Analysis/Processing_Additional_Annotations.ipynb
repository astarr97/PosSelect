```python
mgc = []
astro = [] #ASCNT and ASCT
ogc = []
opc = []
granule = []
l23 = []
pvalb = []
vip = []
sst = []
l4 = []
amy = []
msn = []
cop = []
l6 = []
l5 = []
smc = []
lamp = []
l6b = []
acbgm = []
nearproj = []

c = 0
for index, row in fcat.iterrows():
    c += 1
    if "MGC" in row[1]:
        mgc.append(1)
    else:
        mgc.append(0)
    if "ASCT" in row[1] or "ASCNT" in row[1]:
        astro.append(1)
    else:
        astro.append(0)
    if "OGC" in row[1]:
        ogc.append(1)
    else:
        ogc.append(0)
    if "OPC" in row[1]:
        opc.append(1)
    else:
        opc.append(0)
    if "CBGRC" in row[1]:
        granule.append(1)
    else:
        granule.append(0)
    if "ITL23" in row[1]:
        l23.append(1)
    else:
        l23.append(0)
    if "PVALB" in row[1]:
        pvalb.append(1)
    else:
        pvalb.append(0)
    if "VIP" in row[1]:
        vip.append(1)
    else:
        vip.append(0)
    if "SST" in row[1]:
        sst.append(1)
    else:
        sst.append(0)
    if "ITL4" in row[1] and "ITL45" not in row[1]:
        l4.append(1)
    else:
        l4.append(0)
    if "AMY" in row[1]:
        amy.append(1)
    else:
        amy.append(0)
    if "MSN" in row[1] or "D2-CaB" in row[1] or "D1/2-NAC" in row[1] or "D1-Pu" in row[1] or "D2-Pu" in row[1] or "D1-CaB" in row[1]:
        msn.append(1)
    else:
        msn.append(0)
    if "COP" in row[1]:
        cop.append(1)
    else:
        cop.append(0)
    if "ITL6" in row[1]:
        l6.append(1)
    else:
        l6.append(0)
    if "ITL5" in row[1]:
        l5.append(1)
    else:
        l5.append(0)
    if "LAMP" in row[1]:
        lamp.append(1)
    else:
        lamp.append(0)
    if "SMC" in row[1]:
        smc.append(1)
    else:
        smc.append(0)
    if "L6B" in row[1]:
        l6b.append(1)
    else:
        l6b.append(0)
    if "ACBGM" in row[1]:
        acbgm.append(1)
    else:
        acbgm.append(0)
    if "NP-" in row[1]:
        nearproj.append(1)
    else:
        nearproj.append(0)
    if c % 50000 == 0:
        print(c)
```

```python
fcat["Micro"] = mgc
fcat["Astro"] = astro
fcat["Oligo"] = ogc
fcat["OPC"] = opc
fcat["Granule"] = granule
fcat["L2-3"] = l23
fcat["PVALB"] = pvalb
fcat["VIP"] = vip
fcat["SST"] = sst
fcat["L4"] = l4
fcat["AMY"] = amy
fcat["MSN"] = msn
fcat["COP"] = cop
fcat["L6"] = l6
fcat["L5"] = l5
fcat["SMC"] = smc
fcat["LAMP"] = lamp
fcat["L6B"] = l6b
fcat["ACBGM"] = acbgm
fcat["NearProj"] = nearproj
fcat = fcat.drop(1, axis = 1)

fcat.to_csv("HumChp_NC_Final_Rmdup_CREs_NoHLA_Intersect_catlas_processed.txt", sep = "\t")
```

```python
import random
from collections import Counter

def find_mode_random_tie(strings):
    # Count the frequency of each string
    count = Counter(strings)
    
    # Find the maximum frequency
    max_freq = max(count.values())
    
    # Collect all strings with the maximum frequency
    candidates = [key for key, value in count.items() if value == max_freq]
    
    # Randomly select one of the candidates
    mode = random.choice(candidates)
    
    return mode
```
