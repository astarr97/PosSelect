```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import os
```

```python
df = pd.DataFrame()
ind = 1
for file in os.listdir():
    if "Agoglia" in file and ".txt" in file:
        print(file)
        v = pd.read_csv(file, sep = "\t").set_index("Gene").dropna()
        out = []
        for index, row in v.iterrows():
            out.append(np.sign(row["log2FoldChange"])*norm.ppf(1 - row["pvalue"] / 2))
        v["X"] = out
        v = v[["X"]]
        v.columns = [file.replace("_chpr.txt", "").replace("Agoglia_Fraser_2021_DESeq2_HumChp_Hybrid_CS_", "")]
        if ind:
            df = v
            ind = 0
        else:
            df = df.join(v)
df = df.dropna()
df = df.replace(-np.inf, -9)
df = df.replace(np.inf, 9)
df2 = df.copy()
df.to_csv("Input_Agoglia_ZScore.txt", sep = "\t")
for i in df.columns:
    df[i] = np.repeat(1, df.shape[0])
df.to_csv("Input_Agoglia_SHat.txt", sep = "\t")
```
